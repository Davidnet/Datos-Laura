```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plt.style.use('seaborn-whitegrid')
import seaborn
```

```python
dataset_df = pd.read_csv('./datos.csv')
```

```python
dataset_df.head()
```

```python
dataset_df.describe()
```

```python
dataset_df = dataset_df.dropna()
dataset_df = dataset_df[dataset_df.x1.str.contains("x1") == False]
```

```python
dataset_df = dataset_df.astype('float64')
```

```python
dataset_df.describe()
```

```python
dataset_df['y'] = dataset_df['y'].apply(lambda x: x*100)
```

```python
dataset_df.tail()
```

```python
dataset_df.describe()
```

```python
plt.scatter(dataset_df.x1, dataset_df.x2, marker='x')
plt.xlabel('x1 feature')
plt.ylabel('x2 feature')
```

```python
plt.hist(dataset_df.x1, histtype='stepfilled');
```

```python
plt.hist(dataset_df.x2, color='steelblue');
```

```python
plt.hist2d(dataset_df.x1, dataset_df.x2, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
# Set up the axes with gridspec
fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

main_ax.plot(dataset_df.x1, dataset_df.x2, 'ok', markersize=12, alpha=0.2)

x_hist.hist(dataset_df.x1, 40, histtype='stepfilled', orientation='vertical', color='gray');
y_hist.hist(dataset_df.x2, 40, histtype='stepfilled', orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```

```python
X_train = np.array(dataset_df[["x1", "x2"]])
Y_train = np.array(dataset_df["y"])
```

```python
from sklearn import cross_validation, svm
from sklearn.linear_model import LinearRegression

clf = LinearRegression()
clf.fit(X_train, Y_train)
confidence = clf.score(X_train, Y_train)
print("Linear Regressor has confidence {}".format(confidence))
for k in ['linear', 'poly', 'rbf', 'sigmoid']:
    clf = svm.SVR(kernel=k)
    clf.fit(X_train, Y_train)
    confidence = clf.score(X_train, Y_train)
    print(k, confidence)
```
